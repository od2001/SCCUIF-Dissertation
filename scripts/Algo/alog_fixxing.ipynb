```python

# Convert data to DataFrame for easier manipulation
df = pd.read_json("image_comparison_rev_2.json")

df
```

```python
df = df.replace('IMG_2276', 'B')
df = df.replace('IMG_6533', 'C')
df = df.replace('IMG_6544', 'A')
```

```python
df.sort_values(by="comp_score", ascending=False)
```

```python

# Extract unique image names
images = sorted(set(df['image1'].unique()) | set(df['image2'].unique()))

# Create a mapping from image names to indices
image_index = {image: idx for idx, image in enumerate(images)}

# Initialize matrices
comp_matrix = np.zeros((len(images), len(images)))
occurnences_matrix = np.zeros((len(images), len(images)))
placement_matrix = np.zeros((len(images), len(images)))
colour_matrix = np.zeros((len(images), len(images)))
person_matrix = np.zeros((len(images), len(images)))

# Populate matrices
for _, row in df.iterrows():
    idx1 = image_index[row['image1']]
    idx2 = image_index[row['image2']]
    
    comp_matrix[idx1, idx2] = row['comp_score']
    occurnences_matrix[idx1, idx2] = row['occurnences_score']
    placement_matrix[idx1, idx2] = row['placement_score']
    colour_matrix[idx1, idx2] = row['colour_score']
    person_matrix[idx1, idx2] = row['person_score']

# Function to plot a matrix
def plot_matrix(matrix, title):
    plt.figure(figsize=(10, 8))
    plt.title(title)
    plt.imshow(matrix, cmap='cool', interpolation='nearest')
    plt.colorbar()
    plt.xticks(range(len(images)), images, rotation=90)
    plt.yticks(range(len(images)), images)
    plt.show()

# Plot each matrix
plot_matrix(comp_matrix, "Comparision Score Matrix")
plot_matrix(occurnences_matrix, "Occurences Score Matrix")
plot_matrix(placement_matrix, "Placement Score Matrix")
plot_matrix(colour_matrix, "Colour Score Matrix")
plot_matrix(person_matrix, "Person Score Matrix")
```
