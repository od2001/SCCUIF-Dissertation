# Histo


```python

def histogram(image,per,hue,sat,val):
    total_pixels = image.shape[0] * image.shape[1]

    hue_histo = cv2.calcHist([image],[0], None, [255], [0,179])
    sat_histo = cv2.calcHist([image],[1], None, [255], [0,255])
    val_histo = cv2.calcHist([image],[2], None, [255], [0,255])


    if per == True:
        hue_histo = (hue_histo/total_pixels) * 100
        sat_histo = (sat_histo/total_pixels) * 100
        val_histo = (val_histo/total_pixels) * 100


    if hue == True:
        plt.subplot(131)
        plt.figure(figsize=[20,10])
        plt.plot(hue_histo, color='purple')
        plt.ylabel("Pixel Count(px)" if not per else "Percentage of Pixels(%)")
        plt.xlabel("Intensity Level")
        plt.xlim(0,179)
        plt.xticks(np.arange(0, 179, step=5))
        # plt.xlim(0,2)


    if sat == True:
        plt.subplot(132)
        plt.figure(figsize=[20,20])
        plt.plot(sat_histo, color='cyan')
        plt.ylabel("Pixel Count(px)" if not per else "Percentage of Pixels(%)")
        plt.xlabel("Intensity Level")

    if val == True:
        plt.subplot(133)
        plt.figure(figsize=[40,20])
        plt.plot(val_histo, color='yellow')
        plt.ylabel("Pixel Count(px)" if not per else "Percentage of Pixels(%)")
        plt.xlabel("Intensity Level")
```
